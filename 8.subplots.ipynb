```python
fig, ax= plt.subplots()

ax.scatter(df['avg'],df['strike_rate'],color='red',marker='*',cmap='winter')
#to add title
#previous= plt.title('something')
# in subplots
ax.set_title('graph')

#to label axis 
#previous = plt.xlabel('kcdj')
#in sublots
ax.set_xlabel('xaxis')
ax.set_ylabel('yaxis')

#plt.show()
fig.show()
```

```python
#plotting two plots together
plt.subplots(nrows=2,ncols=2)
# 2*1=2 subplots
#norws=2,ncols=2 . 2*2=4 subplots
```

```python
fig,ax= plt.subplots(nrows=2,ncols=1)
#fatching 1st axis
ax[0]
#fatching 2nd axis
ax[1]
```

```python
fig,ax= plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))

ax[0].scatter(df10['avg'],df10['strike_rate'])
ax[1].scatter(df10['avg'],df10['runs'])

ax[0].set_title('avg vs str')
ax[0].set_ylabel('str')

ax[1].set_title('avg vs runs')
ax[1].set_ylabel('runs')
ax[1].set_xlabel('avg')
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2)

ax[0,0].scatter(df10['avg'],df10['strike_rate'])
ax[0,1].scatter(df10['avg'],df10['runs'])
ax[1,0].hist(df10['runs'])
ax[1,1].hist(df10['strike_rate'])
```
