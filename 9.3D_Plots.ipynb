```python
#3d Line Plot
fig=plt.figure()
ax= plt.subplot(projection='3d')
x=[0,1,5]
y=[0,10,13]
z=[0,13,20]

ax.scatter3D(x,y,z,s=[100,100,100],color='red')#markersize nhi hai syd
ax.plot3D(x,y,z)
```

```python
#3d Surface Plot

x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)

 
```

```python
xx,yy=np.meshgrid(x,y)
z=xx**2+yy**2
```

```python
fig=plt.figure(figsize=(6,4))
ax=plt.subplot(projection='3d')
ax.plot_surface(xx,yy,z,cmap='viridis')
```

```python
#to make colorbar 
# store fun in variable and then pass
fig=plt.figure(figsize=(6,4))
ax=plt.subplot(projection='3d')
bar=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(bar)
```

```python
#ex2
z=np.sin(xx) + np.cos(yy)
fig=plt.figure(figsize=(6,4))
ax=plt.subplot(projection='3d')
bar=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(bar)
```

```python
#contour graph
#convert 3d to 2d graph
```

```python
fig=plt.figure(figsize=(6,4))
ax=plt.subplot()
bar=ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(bar)
```
