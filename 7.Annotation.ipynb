```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Annotation- giving points name on graph
Horizontal and Vertical line

```python
#plt.text - giving points name on graph

plt.figure(figsize=(3,3))

x=[1,2,3,4]
y=[5,6,7,8]
plt.scatter(x,y)

plt.text(1,5,'point1',color='red')
plt.text(2,6,'point2',color='blue')
plt.text(3,7,'point3',color='green')
plt.text(4,8,'point4',color='orange',size=20)
```

```python
#on dataset
df=pd.read_csv(r'C:\Users\HP\Desktop\all folder\matplot\matplot\batter.csv')
df50=df.sample(50,random_state=5)
df50.shape
```

```python
# to name on multiple points loop  + plt.text

plt.figure(figsize=(9,4))
plt.scatter(df50['avg'],df50['strike_rate'],s=df50['runs'])
#now here to name points we will run loop
for i in range(df50.shape[0]):
    plt.text(df50['avg'].values[i],df50['strike_rate'].values[i],df50['batter'].values[i],size=8)

plt.show()    
```

Horizontal and Vertical line
-plt.axhline()
-plt.axvline()

```python
plt.figure(figsize=(9,4))
plt.scatter(df50['avg'],df50['strike_rate'],s=df50['runs'])
for i in range(df50.shape[0]):
    plt.text(df50['avg'].values[i],df50['strike_rate'].values[i],df50['batter'].values[i],size=8)

plt.axhline(130,color='red')
plt.axhline(150,color='green')
plt.axvline(30,color='red')
```
