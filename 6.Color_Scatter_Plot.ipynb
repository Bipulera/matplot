```python
#import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Colored Scatter Plots

```python
iris=pd.read_csv(r'C:\Users\HP\Desktop\all folder\matplot\matplot\iris.csv')
iris
```

```python
#making scatter plot of iris
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('sepal_length')
plt.ylabel('petal_length')
```

```python
iris['Species'].value_counts()
```

```python
#replace 
iris['Species']=iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
iris.head(1)
```

```python
# c -> cmap(summer/winter/jet) -> alpha(transparency 0-1)
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.7)
```

```python
#colorbar
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.colorbar()
```

```python
#plot size --plt.figure(figsize(x,y))
plt.figure(figsize=(3,3))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'],c=iris['Species'],cmap='winter')
plt.colorbar()
```
