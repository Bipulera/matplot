```python
#from a pandas df
batsman=pd.read_csv(r'C:\Users\HP\Desktop\all folder\matplot\matplot\sharma-kohli.csv')
batsman
```

```python
#plotting one graph
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
#plotting multiple graphs
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
#label title
# xlabel
# ylabel
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.plot(batsman['index'],batsman['V Kohli'])

plt.title('Rohit and virat carrer stats')
plt.xlabel('years')
plt.ylabel('runs')
```

```python
# color(hex) and line(width and style) and marker(size)
plt.plot(batsman['index'],batsman['RG Sharma'], color='black', linestyle='dotted', linewidth=1, marker='D')
plt.plot(batsman['index'],batsman['V Kohli'], color='red' , linestyle='dashed', linewidth=2 , marker='+')

plt.title('Rohit and virat carrer stats')
plt.xlabel('years')
plt.ylabel('runs')
```

```python
#legend -> location 
#legend means labeling lines

#syntax is ( label ) + plt.legend()

plt.plot(batsman['index'],batsman['RG Sharma'], color='black', label='rohit' )
plt.plot(batsman['index'],batsman['V Kohli'], color='red' , label='kohli')
plt.title('Rohit and virat carrer stats')
plt.xlabel('years')
plt.ylabel('runs')

plt.legend(loc='upper center')

#location -> loc
#default is plt.legend(loc='best') gap find on its own
# upper right/left , lower right/left, center left/right, lower/upper center, center
```

```python
#limiting axes range-> plt.ylim(range)
# works for outlier.

price = [48000,54000,57000,49000,47000,45000,10000000]
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year,price)
```

```python
#as here we cannot see variation of starting years
price = [48000,54000,57000,49000,47000,45000,1000000]
year = [2015,2016,2017,2018,2019,2020,2021]
#plt.plot(x,y)
plt.plot(year,price)
plt.ylim(0,70000)
#plt.xlim(2017,2020)
```

```python
#grid
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
#plt.plot(x,y)
plt.plot(year,price)
plt.grid()
```

```python
#show- in other text editor graph will not show until plt.show()

price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
#plt.plot(x,y)
plt.plot(year,price)
plt.show()
```
